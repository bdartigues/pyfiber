```python
def interpolate(sampling_rate,window,time,data):
    new_time = np.linspace(window[0],window[1],sampling_rate*(window[1]-window[0]))
    return new_time,np.interp(new_time,time,data)

def interpolate_average(sampling_rate,window,time_series,data_series,mean_sampling=False):
    interpolated_data_series = []
    new_time = np.linspace(window[0],window[1],sampling_rate*(window[1]-window[0]))
    for a,b in zip(time_series,data_series):
        interpolated_data_series.append(np.interp(new_time,a,b))
    mean_data = sum(interpolated_data_series)/len(interpolated_data_series)
    if mean_sampling:
        mean_time = np.linspace(window[0],window[1],mean_sampling*(window[1]-window[0]))
        mean_data = np.interp(mean_time,new_time,mean_data)
    else:
        mean_time = new_time
    return interpolated_data_series,mean_data,new_time,mean_time

def plot_data_and_mean(sampling_rate,window,time_series,data_series,mean_sampling=False,**kwargs):
    cfg = {'figsize':(20,10),'c':'k','linewidth':1,'alpha':0.3}
    for key in cfg.keys():
        if key in kwargs.keys():
            cfg[key] = kwargs[key]
    plt.figure(figsize=(cfg['figsize']))
    data,mean_data,time,mean_time = interpolate_average(sampling_rate,window,time_series,data_series,mean_sampling)
    for n,z in enumerate(data):
        plt.plot(time,z,alpha=cfg['alpha'])
    plt.plot(mean_time,mean_data,c='k')
    plt.axvline(0,alpha=1,linewidth=cfg['linewidth'],c=cfg['c'])
    return kwargs
```
