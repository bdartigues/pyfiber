```python
def create_folder(df):
    rat = df.loc[:,'Rat'].values[0]
    exp = df.loc[:,'Experiment'].values[0]
    for i in df.index:
        session = df.loc[i,'Session Number']
        typ = ''.join((''.join(df.loc[i,'Recording Type'].split('.'))).split(' '))
        #os.system(f'mkdir {exp}_rat_{rat}_j{session}')
        print(f'mkdir {exp}_rat_{rat}_{typ}_j{session}')
```
