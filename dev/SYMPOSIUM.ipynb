```python
for thing in a.__dict__.keys():
    if 'pre' in thing:
        name = thing.split('pre')[1]
        ai = a.__dict__[thing]
        bi = a.__dict__['post'+name]
        if type(ai[0]) == np.ndarray:
            ai = [np.mean(i) for i in ai]
            bi = [np.mean(i) for i in bi]
        t,p = stats.ttest_ind(ai,bi)
        message = ''
        if p < 0.05:
            message = 'youhou!!'
        if p < 0.01:
            message = 'double youhou!!'
        ci,di = stats.norm.interval(0.68, loc=np.mean(ai), scale=np.array(ai).std()/np.sqrt(len(ai)))
        interval = '['+str(round(ci,4))+','+str(round(di,4))+']'
        print(f"{name:<10}: CI={interval:<21} t={t:<25} ; p={p:<25} {message}")
```
